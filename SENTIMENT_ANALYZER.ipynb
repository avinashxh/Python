```python
from transformers import pipeline
```

```python
senti_pipeline = pipeline("sentiment-analysis")
```

```python
senti_pipeline("im happy that you are here")
```

```python
senti_pipeline("hi,im fine but these days i was thinking of suicide")
```

```python
senti_pipeline("my sleep cycle is good")
```

```python
senti_pipeline("it was a hectic day")
```

```python
senti_pipeline("shradha is a good girl")
```

```python
selected_row = 3
column1 = 'Hows your study life going ?'
column2 = 'Hows your life recently ?'
concatenated_column = df[column1] + ' ' + df[column2]
print(concatenated_column)
cell_value = concatenated_column[selected_row]
print(cell_value)
b = senti_pipeline(cell_value)
print(b)
```

```python
selected_rows = [1, 2, 3, 4, 5, 6, 7, 8]
column1 = 'Hows your study life going ?'
column2 = 'Hows your life recently ?'
column3 = 'Are you going happy or unhappy these days ?'

for row in selected_rows:
    concatenated_column = df[column1][row] + ' --- ' + df[column2][row] + ' --- ' +df[column3][row]
    print(concatenated_column)
    b = senti_pipeline(concatenated_column)
    print(f"Row {row}: {b}")
```
